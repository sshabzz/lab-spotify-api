```python
from config import *
import spotipy
import json
from spotipy.oauth2 import SpotifyClientCredentials
import numpy as np

sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=client_id,
                                                           client_secret=client_secret_id))
```

```python
import pandas as pd
hot = pd.read_csv('hot100.csv')
not_hot = pd.read_csv('not_hot.csv')

display(hot.head())
not_hot.head()
```

```python
song = sp.search(q="tracks: Kill Bill artist: SZA", limit=1) 
song['tracks']['items'][0]['id']
```

```python
#function to search for songs ids:
def search_song(title,artist):
    try: 
        string = "tracks: " + title + "artist: " + artist
        song=sp.search(q=string,limit=1)
        id = song['tracks']['items'][0]['id']
    except:
        #print("Song not found!")
        id = np.nan
    
    return id
```

```python
hot2 = hot.copy()
hot2['id'] = list(map(search_song,hot2['Title'],hot2['Artist']))

hot2.head()
```

```python
from time import sleep
from random import randint
```

```python
#splitting not_hot songs into chunks:
def chunks(lst, n):
    for i in range(0, len(lst), n):
        yield lst[i:i + n]
```

```python
split_title = list(chunks(hot['Title'], 50))
split_artist = list(chunks(hot['Artist'], 50))
```

```python
split_title_not = list(chunks(not_hot['Title'], 100))
split_artist_not = list(chunks(not_hot['Artist'], 100))
```

```python
#extracting the list of hot ids:
hot_id=[]
for i in range(len(split_title)):
    for j in range(0,50):
        hot_id.append(search_song(list(split_title[i])[j],list(split_artist[i])[j]))
    sleep(10)

hot_id
```

```python
#extracting the list of not_hot ids
not_hot_id = []

for i in range(len(split_title_not)):
    for j in range(0,100):
        not_hot_id.append(search_song(list(split_title_not[i])[j],list(split_artist_not[i])[j]))
    sleep(10)
```

```python
not_hot['id'] = not_hot_id

not_hot.head()
```

```python
#dropping the missing values:
not_hot['id'].isna().sum()
```

```python
not_hot.dropna(inplace=True)
not_hot.shape
```

```python
#appending not_hot ids to not_hot df:
not_hot_id = list(not_hot['id'])
```

```python
#splitting not_hot ids into chunks:
split_not_hot_id = list(chunks(not_hot_id, 100))
```

```python
#function to extract audio features:
def get_audio_features(list_of_songs_ids):
    features = sp.audio_features(list_of_songs_ids)
    return features

hot_features = get_audio_features(hot_id)
```

```python
#extracting not_hot features:
not_hot_features=[]

for i in range(len(split_not_hot_id)):
    not_hot_features.append(get_audio_features(list(split_not_hot_id[i])))
    sleep(10)
```

```python
#gettig the hot features:
hot_ft = pd.DataFrame(data=[hot_features[i].values() for i in range(len(hot_features))],columns=hot_features[0].keys())
hot_ft.head()
```

```python
#concatenationg hot and not_hot df:
hot_all = pd.concat([hot,hot_ft],axis=1)
display(hot_all.shape)
display(hot_all.isna().sum())
hot_all.head()
```
